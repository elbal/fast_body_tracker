```python
from pathlib import Path

import h5py
import pandas as pd
```

```python
file_path = Path("../data/test_run/body.h5")
```

```python
with h5py.File(file_path, "r") as f:
    ts_df = pd.DataFrame(
        {
            "frame_idx": f["ts"]["frame_idx"][:],
            "ts": f["ts"]["ts"][:],
            "system_ts": f["ts"]["system_ts"][:],
        }
    )

    contribution_size = f["joints"][next(iter(f["joints"]))]["contributions"].shape[1]

    body_dfs = {
        body_key: pd.DataFrame(
            {
                "frame_idx": f["joints"][body_key]["frame_idx"][:],
                "tags": f["joints"][body_key]["tags"][:],
                **{
                    f"confidence_{i}": f["joints"][body_key]["confidences"][:, i]
                    for i in range(32)
                },
                **{
                    f"contribution_{i}": f["joints"][body_key]["contributions"][:, i]
                    for i in range(contribution_size)
                },
            }
        )
        for body_key in f["joints"].keys()
    }
```
